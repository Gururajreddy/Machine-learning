4. OneHotEncoding with Top Categories

```python
counts = df['brand'].value_counts()
```

```python
df['brand'].nunique()
threshold = 100
```

```python
rep=counts[counts<=threshold].index
```

```python
pd.get_dummies(df['brand'].replace(rep, 'uncommon')).sample(5)
```
